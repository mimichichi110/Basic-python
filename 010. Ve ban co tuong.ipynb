```python
import turtle
tl = turtle.Turtle()
tl.pensize(2)
tl.speed(5)
d = 50
tl.penup()
tl.goto(-4 * d, 4 * d + d/2)
tl.pendown()

# Khung bên trong
i = 0
while i < 2:
    tl. forward(8*d)
    tl.right(90)
    tl.forward(9*d)
    tl.right(90)
    i += 1

# 8 đường ngang
j = 0
while j < 8:
    tl.right(90)
    tl.forward(d)
    tl.left(90)
    tl.forward(8*d)
    tl.penup()
    tl.backward(8*d)
    tl.pendown()
    j += 1

# 4 đường dọc phía trên
tl.penup()
tl.goto(-3 * d, 4 * d + d/2)
tl.pendown()

k = 0
while k < 7:
    tl.right(90)
    tl.forward(4 * d)
    tl.penup()
    tl.backward(4 * d)
    tl.left(90)
    tl.forward(d)
    tl.pendown()
    k += 1

# 4 đường dọc phía dưới
tl.penup()
tl.goto(-3 * d, -d/2)
tl.pendown()

l = 0
while l < 7:
    tl.right(90)
    tl.forward(4 * d)
    tl.penup()
    tl.backward(4 * d)
    tl.left(90)
    tl.forward(d)
    tl.pendown()
    l += 1

# Ô chéo phía trên
tl.penup()
tl.goto(-d, 4 * d + d/2)
tl.pendown()
tl.right(45)
tl.forward(((d**2 + d**2)**0.5)*2)

tl.penup()
tl.goto(d, 4 * d + d/2)
tl.pendown()
tl.right(90)
tl.forward(((d**2 + d**2)**0.5)*2)

# Ô chéo phía dưới
tl.penup()
tl.goto(-d, -2 * d - d/2)
tl.pendown()
tl.left(90)
tl.forward(((d**2 + d**2)**0.5)*2)

tl.penup()
tl.goto(d, -2 * d - d/2)
tl.pendown()
tl.right(90)
tl.forward(((d**2 + d**2)**0.5)*2)

# Khung đậm
du = 6

tl.penup()
tl.right(45)
tl.goto(-4*d - du, -4 * d -d/2 -du)
tl.pendown()
tl.pensize(5)

m = 0
while m < 2:
    tl.right(90)
    tl.forward(9 * d + 2 * du)
    tl.right(90)
    tl.forward(8 * d + 2 * du)
    m += 1

#Khung ngoài cùng
tl.pensize(2)
tl.penup()
tl.goto(-4*d - d/2, -4*d - d/2 - d/2)
tl.pendown()

n = 0
while n < 2:
    tl.right(90)
    tl.forward(9 * d + 2 * d/2)
    tl.right(90)
    tl.forward(8 * d + 2 * d/2)
    n += 1

#Tuyến tốt phía trên: góc vuông trên bên phải
tl.left(90)
tl.penup()
tl.goto(-4*d +3, d/2 + d + d/5 + 3)
tl.pendown()
o = 0
while o < 4:
    tl.forward(d/5)
    tl.left(90)
    tl.forward(d/5)
    
    tl.penup()
    tl.forward(d + 3 + (d-d/5-3))
    tl.right(90)
    tl.backward(d/5)
    tl.pendown()
    o += 1

#Tuyến tốt phía trên: góc vuông dưới bên phải
tl.penup()
tl.goto(-4*d +3, d/2 + d - d/5 - 3)
tl.pendown()

a = 0
while a < 4:
    tl.backward(d/5)
    tl.left(90)
    tl.forward(d/5)
    
    tl.penup()
    tl.forward(d + 3 + (d-d/5-3))
    tl.right(90)
    tl.forward(d/5)
    tl.pendown()
    a += 1

#Tuyến tốt phía trên: góc vuông trên bên trái trên
tl.penup()
tl.goto(-2*d - d/5 - 3, d/2 + d + 3)
tl.left(90)
tl.pendown()

b = 0
while b < 4:
    tl.forward(d/5)
    tl.left(90)
    tl.forward(d/5)
    tl.right(90)
    
    tl.penup()
    tl.forward(d + 3 + (d-d/5-3))
    tl.left(90)
    tl.backward(d/5)
    tl.right(90)
    tl.pendown()
    b += 1

#Tuyến tốt phía trên: góc vuông trên bên trái dưới
tl.penup()
tl.goto(-2*d - 3 - d/5, d/2 + d -3)
tl.pendown()

c = 0
while c < 4:
    tl.forward(d/5)
    tl.right(90)
    tl.forward(d/5)
    tl.left(90)
    
    tl.penup()
    tl.forward(3 + d +(d-3-d/5))
    tl.left(90)
    tl.forward(d/5)
    tl.right(90)
    tl.pendown()
    c += 1

#Tuyến tốt phía dưới: góc vuông trên bên phải
tl.right(90)
tl.penup()
tl.goto(-4*d +3, -d/2 - (d-d/5-3))
tl.pendown()
e = 0
while e < 4:
    tl.forward(d/5)
    tl.left(90)
    tl.forward(d/5)
    
    tl.penup()
    tl.forward(d + 3 + (d-d/5-3))
    tl.right(90)
    tl.backward(d/5)
    tl.pendown()
    e += 1

#Tuyến tốt phía dưới: góc vuông dưới bên phải
tl.penup()
tl.goto(-4*d +3, -d/2 - d - (d/5+3))
tl.pendown()

f = 0
while f < 4:
    tl.backward(d/5)
    tl.left(90)
    tl.forward(d/5)
    
    tl.penup()
    tl.forward(d + 3 + (d-d/5-3))
    tl.right(90)
    tl.forward(d/5)
    tl.pendown()
    f += 1
    
#Tuyến tốt phía dưới: góc vuông bên trái trên
tl.penup()
tl.goto(-2*d - d/5 - 3, -d/2 - (d - 3))
tl.left(90)
tl.pendown()

g = 0
while g < 4:
    tl.forward(d/5)
    tl.left(90)
    tl.forward(d/5)
    tl.right(90)
    
    tl.penup()
    tl.forward(d + 3 + (d-d/5-3))
    tl.left(90)
    tl.backward(d/5)
    tl.right(90)
    tl.pendown()
    g += 1
    
#Tuyến tốt phía dưới: góc vuông bên trái dưới
tl.penup()
tl.goto(-2*d - 3 - d/5, -d/2 - d -3)
tl.pendown()

h = 0
while h < 4:
    tl.forward(d/5)
    tl.right(90)
    tl.forward(d/5)
    tl.left(90)
    
    tl.penup()
    tl.forward(3 + d +(d-3-d/5))
    tl.left(90)
    tl.forward(d/5)
    tl.right(90)
    tl.pendown()
    h += 1
    
#Tuyến pháo trên: góc vuông trên bên phải
tl.penup()
tl.goto(-2*d - (d-3), d/2 + 2*d + 3 + d/5)
tl.right(90)
tl.pendown()

p = 0
while p < 2:
    tl.forward(d/5)
    tl.left(90)
    tl.forward(d/5)
    
    tl.penup()
    tl.forward(5*d + 3 + (d-d/5-3))
    tl.right(90)
    tl.backward(d/5)
    tl.pendown()
    p += 1

#Tuyến pháo trên: góc vuông dưới bên phải
tl.penup()
tl.goto(-2*d - (d-3), d/2 + d + (d- d/5 - 3))
tl.pendown()
q = 0
while q < 2:
    tl.backward(d/5)
    tl.left(90)
    tl.forward(d/5)
    
    tl.penup()
    tl.forward(5*d + 3 + (d-d/5-3))
    tl.right(90)
    tl.forward(d/5)
    tl.pendown()
    q += 1

#Tuyến pháo trên: góc vuông trên bên trái
tl.penup()
tl.goto(-3*d - d/5 -3, d/2 + 2*d +3)
tl.left(90)
tl.pendown()

r = 0
while r < 2:
    tl.forward(d/5)
    tl.left(90)
    tl.forward(d/5)
    tl.right(90)
    
    tl.penup()
    tl.forward(5*d + 3 + (d-d/5-3))
    tl.left(90)
    tl.backward(d/5)
    tl.right(90)
    tl.pendown()
    r += 1

#Tuyến pháo trên: góc vuông dưới bên trái
tl.penup()
tl.goto(-3*d - d/5 -3, d/2 + d + (d-3))
tl.pendown()

s = 0
while s < 2:
    tl.forward(d/5)
    tl.right(90)
    tl.forward(d/5)
    tl.left(90)
    
    tl.penup()
    tl.forward(5*d + 3 + (d-d/5-3))
    tl.left(90)
    tl.forward(d/5)
    tl.right(90)
    tl.pendown()
    s += 1

#Tuyến pháo dưới: góc vuông trên bên phải
tl.penup()
tl.goto(-2*d - (d-3), -d/2 - d - (d - 3 - d/5))
tl.right(90)
tl.pendown()

t = 0
while t < 2:
    tl.forward(d/5)
    tl.left(90)
    tl.forward(d/5)
    
    tl.penup()
    tl.forward(5*d + 3 + (d-d/5-3))
    tl.right(90)
    tl.backward(d/5)
    tl.pendown()
    t += 1
    
#Tuyến pháo dưới: góc vuông dưới bên phải
tl.penup()
tl.goto(-2*d - (d-3), -d/2 - 2*d - d/5 - 3)
tl.pendown()
u = 0
while u < 2:
    tl.backward(d/5)
    tl.left(90)
    tl.forward(d/5)
    
    tl.penup()
    tl.forward(5*d + 3 + (d-d/5-3))
    tl.right(90)
    tl.forward(d/5)
    tl.pendown()
    u += 1

#Tuyến pháo dưới: góc vuông trên bên trái
tl.penup()
tl.goto(-3*d - d/5 -3, -d/2 - d - (d-3))
tl.left(90)
tl.pendown()

v = 0
while v < 2:
    tl.forward(d/5)
    tl.left(90)
    tl.forward(d/5)
    tl.right(90)
    
    tl.penup()
    tl.forward(5*d + 3 + (d-d/5-3))
    tl.left(90)
    tl.backward(d/5)
    tl.right(90)
    tl.pendown()
    v += 1

#Tuyến pháo dưới: góc vuông dưới bên trái
tl.penup()
tl.goto(-3*d - d/5 -3, -d/2 - 2*d -3)
tl.pendown()

w = 0
while w < 2:
    tl.forward(d/5)
    tl.right(90)
    tl.forward(d/5)
    tl.left(90)
    
    tl.penup()
    tl.forward(5*d + 3 + (d-d/5-3))
    tl.left(90)
    tl.forward(d/5)
    tl.right(90)
    tl.pendown()
    w += 1
    
turtle.done()
```

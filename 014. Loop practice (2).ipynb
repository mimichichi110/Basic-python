```python
# Nhập số tự nhiên N và in ra hình dấu * theo quy luật. Không dùng toán tử nhân

N = int(input("Nhập N:"))

for i in range(N):
    for j in range(N-i):
        print("*",end='')
    print('')
```

```python
# Nhập số tự nhiên N và in ra hình dấu * theo quy luật. Không dùng toán tử nhân

N = int(input("Nhập N:"))

for i in range(N-1,-1,-1):
    for j in range(N-i):
        print("*",end='')
    print('')
```

```python
# Nhập số tự nhiên N và in ra hình dấu * theo quy luật

N = int(input("Nhập N:"))

for i in range(2*N-1):
    if i <= N-1:
        print("*"*(i+1))
    else:
        print("*"*(2*N -1 - i))
```

```python
# Nhập số tự nhiên N và in ra hình dấu * theo quy luật

N = int(input("Nhập N:"))

for i in range(N):
    if i <= N/2:
        print(" "*int((N-2*i)/2),"*"*(2*i+1))
    else:
        print(" "*int((2*i-N+1)/2),"*"*(2*N-2*i-1))
```

```python
# Tối đa bao nhiêu cách đổi tờ tiền 100k ra mệnh giá 2k, 3k, 5k

n = 100 #k VND
a = 2 #k VND
b = 3 #k VND
c = 5 #k VND
count = 0
print('\033[1m'+"Cách 2k   3k   5k"+'\033[0m')
for i in range(n // a + 1):
    for j in range(n // b + 1):
        for k in range(n // c + 1):
            if a*i + b*j + c*k == 100:
                count += 1
                print("{0}    {1}    {2}    {3}".format(count, i, j, k))
print("Tổng số cách:", count)               
```

```python
from random import randint

while True:
    print("Gõ vào Búa/Kéo/Bao hoặc Q để dừng chơi: ")
    user = input().lower()
    if user == "q":
        break
        
    computer = randint(0,2)

    if computer == 0:
        computer = "Búa"
    elif computer == 1:
        computer = "Bao"
    elif computer == 2:
        computer = "Kéo"

    print("--")
    print("Bạn chọn: " + str(user))
    print("Máy tính chọn: " + str(computer))
    print("--")    

    if user == computer:
        print("Hòa")
    else:
        if user == "keo" or "kéo":
            if computer == "bua":
                print("Bạn thua!")
            else:
                print("Bạn thắng!")

        elif user == "bua" or "búa":
            if computer == "keo":
                print("Bạn thắng!")
            else:
                print("Bạn thua!")

        elif user == "bao":
            if computer == "keo":
                print("Bạn thua!")
            else:
                print("Bạn thắng!")
        else:
            print("Nhập sai")

print("Bạn đã dừng chơi")
```

```python
from tkinter import *
import random

root = Tk()
root.geometry("400x400")
root.title("Trò chơi Búa Kéo Bao")

computer_value = {"0":"Búa", "1":"Kéo", "2":"Bao"}

def reset_game():
    bua_btn['state'] = 'active'
    keo_btn['state'] = 'active'
    bao_btn['state'] = 'active'
    
def button_disable():
    bua_btn['state'] = 'disable'
    keo_btn['state'] = 'disable'
    bao_btn['state'] = 'disable'
    
def isrock():
    computer = computer_value[str(random.randint(0,2))]
    if computer == "Búa":
        result = "Hòa!"
    elif computer == "Kéo":
        result = "Bạn thắng!"
    else:
        result = "Máy tính thắng!"
    lbResult['text'] = result
    lbPlayer['text'] = "Búa"
    lbComputer['text'] = computer
    button_disable()
        
def is_scissors():
    computer = computer_value[str(random.randint(0,2))]
    if computer == "Búa":
        result = "Máy tính thắng!"
    elif computer == "Kéo":
        result = "Hòa!"
    else:
        result = "Bạn thắng!"
    lbResult['text'] = result
    lbPlayer['text'] = "Kéo"
    lbComputer['text'] = computer
    button_disable()
        
def is_paper():
    computer = computer_value[str(random.randint(0,2))]
    if computer == "Búa":
        result = "Bạn thắng!"
    elif computer == "Kéo":
        result = "Máy tính thắng!"
    else:
        result = "Hòa!"
    lbResult['text'] = result
    lbPlayer['text'] = "Bao"
    lbComputer['text'] = computer
    button_disable()


name = Label(root, text = "Trò chơi Búa Kéo Bao", font = "normal 20 bold", fg = "green").pack(pady = 20)

l1 = Label(root, text = "Player", font = 13)
l2 = Label(root, text = "VS", font = 13)
l3 = Label(root, text = "Computer", font = 13)
l1.place(x=20, y=100)
l2.place(x=180, y=100)
l3.place(x=280, y=100)

# result = Entry(root, text = '',font = "normal 20 bold", width=15)
# result.place(x=80, y=180)

# Thêm các nút bấm thao tác
photo_bua = PhotoImage(file=r'bua.png')
bua_btn = Button(root, text="",image=photo_bua,compound=LEFT, command = isrock)
bua_btn.place(x=20, y=250)

photo_keo = PhotoImage(file=r'keo.png')
keo_btn = Button(root, text="",image=photo_keo,compound=LEFT, command = is_scissors)
keo_btn.place(x=165, y=250)

photo_bao = PhotoImage(file=r'bao.png')
bao_btn = Button(root, text="",image=photo_bao,compound=LEFT, command = is_paper)
bao_btn.place(x=300, y=250)

reset_btn = Button(root, text="Reset", font = "normal 16 bold",command = reset_game)
reset_btn.place(x=165, y=350)

lbPlayer = Label(root, text= "",font = "normal 15 bold")
lbPlayer.place(x=20,y=130)
lbComputer = Label(root, text= "",font = "normal 15 bold")
lbComputer.place(x=300,y=130)
lbResult = Label(root, text = "", font = "normal 20 bold", justify='center')
lbResult.place(x=100,y=190)

root.mainloop()
```

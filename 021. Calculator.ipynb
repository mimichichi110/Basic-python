```python
from tkinter import *
import math

text_display = ""
first_number = 0
second_number = 0
action = ''
float_pressed = False

def number_press(event):
    global text_display
    pressed = event.widget['text']
    if text_display == '' or text_display == '0':
        if pressed == '0':
            text_display = '0'
        else:
            if text_display == '0':
                text_display = ''
            text_display += pressed
    else:
        if text_display == '0':
            text_display = ''
        text_display += pressed    
    screen.delete(0,'end')
    screen.insert(0, text_display)

def action_press(event):
    global first_number
    first_number = float(screen.get())
    global action
    action = event.widget['text']
    global text_display
    text_display = ''

def finish_press(event):
    global second_number
    second_number = float(screen.get())
    
    if action == '+':
        result = first_number + second_number
    elif action == '-':
        result = first_number - second_number
    elif action == '*':
        result = round(first_number * second_number,2)
    else:
        if second_number != 0:
            result = first_number / second_number
        else:
            result = 'NaN'
    screen.delete(0,'end')
    screen.insert(0,result)
    
def clear_press(event):
    global text_display
    text_display = ''
    global first_number
    first_number = 0
    global second_number
    second_number = 0
    global action
    action = ''
    global float_pressed
    float_pressed = False
    screen.delete(0,'end')
    screen.insert(0,'0')

def monoaction_press(event):
    global first_number
    first_number = float(screen.get())
    global action
    action = event.widget['text']
    if action == '1/x':
        if first_number != 0:
            result = 1/first_number
        else:
            result = "NaN"
    elif action == 'sqrt':
        result = math.sqrt(first_number)
    elif action == '%':
        result = first_number * 100
    global text_display
    text_display = ''
    screen.delete(0,'end')
    screen.insert(0,result)    

def reverse_press(event):
    global first_number
    first_number = float(screen.get())
    global action
    action = event.widget['text']
    if action == '+/-':
        if first_number != 0:
            result = - first_number
        else:
            result = 0
    global text_display
    text_display = ''
    screen.delete(0,'end')
    screen.insert(0,result)
def float_press(event):
    global float_pressed
    global text_display
    if not float_pressed:
        text_display += '.'
    float_pressed = True
    
root = Tk()
root.title('Calculator')
root.geometry('305x390')
root.resizable(0, 0)
root.eval('tk::PlaceWindow . center')

screen = Entry(root, font=("Helvetica", 17), justify= 'right')
screen.grid(row=0, column=0, columnspan = 5, padx=5, pady=5, ipadx=10, ipady=10)
screen.insert(0, '0')

keyboards = {'MC': 0, 'MR': 0, 'MS': 0, 'M+': 0, 'M-': 0, '<-': 0, 'CE': 0, 'C': 4, '+/-': 5, 'sqrt': 5, '7': 1, '8': 1,'9': 1, '/': 2, '%': 5, '4': 1, '5': 1, '6': 1, '*': 2, '1/x': 5, '1': 1, '2': 1, '3': 1, '-': 2, '=': 3, '0': 1, '': 0, '.': 7, '+': 2}

# gồm có 30 phím, vậy i chạy từ 0-29
keys = list(keyboards.keys())
row = 1
column = 0
for i in range(len(keys)):
    if i%5==0 and i!=0:
        row += 1
    if i == 26:
        continue
    #phím 0
    if i==25:
        btn = Button(root, text = keys[i], height=2, width=13,font=("Helvetica", 10, 'bold'))
        btn.grid(row=row, column=i%5, padx=5, pady=5, columnspan=2)
    elif i==24:
        btn = Button(root, text = keys[i], height=5, width=5,font=("Helvetica", 10, 'bold'))
        btn.grid(row=row, column=i%5, padx=5, pady=5, rowspan=2, ipady=3)
    else:
        btn = Button(root, text = keys[i], font=("Helvetica", 10, 'bold'), height=2, width=5)
        btn.grid(row=row, column=i%5, padx=5, pady=5)

    if keyboards[keys[i]]==1:
        btn.bind("<Button-1>", number_press)
    elif keyboards[keys[i]]== 2:
        btn.bind("<Button-1>", action_press)
    elif keyboards[keys[i]]== 3:
        btn.bind("<Button-1>", finish_press)
    elif keyboards[keys[i]]== 4:
        btn.bind("<Button-1>", clear_press)
    elif keyboards[keys[i]]== 5:
        btn.bind("<Button-1>", monoaction_press)
    elif keyboards[keys[i]]== 6:
        btn.bind("<Button-1>", reverse_press)
    elif keyboards[keys[i]]== 7:
        btn.bind("<Button-1>", float_press)
root.mainloop()
```

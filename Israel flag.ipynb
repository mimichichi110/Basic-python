```python
import turtle

tl = turtle.Turtle()
tl.shape('turtle')
tl.hideturtle()
tl.pensize(4)
tl.pencolor('blue')

tl.fillcolor('blue')
tl.begin_fill()
tl.forward(300)
tl.right(90)
tl.forward(40)
tl.right(90)
tl.forward(300)
tl.right(90)
tl.forward(40)
tl.end_fill()

tl.penup()
tl.forward(200)
tl.right(90)
tl.pendown()

tl.fillcolor('blue')
tl.begin_fill()
tl.forward(300)
tl.right(90)
tl.forward(40)
tl.right(90)
tl.forward(300)
tl.right(90)
tl.forward(40)
tl.end_fill()

tl.penup()
tl.backward(80)
tl.right(90)
tl.forward(100)
tl.pendown()

tl.forward(100)
tl.right(120)
tl.forward(100)
tl.right(120)
tl.forward(100)

tl.right(30)
tl.penup()
tl.forward(20)
tl.right(90)
tl.forward(50)
tl.right(60)
tl.pendown()

tl.forward(100)
tl.right(120)
tl.forward(100)
tl.right(120)
tl.forward(100)

turtle.done
```

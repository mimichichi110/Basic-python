```python
class Answer:
    
    def __init__(self, text, iscorrect):
        self.text = text
        self.iscorrect = iscorrect
     
    def display(self):
        print("Đáp án: ", self.text, end='  ')
        print(self.iscorrect)

    def clone(self):
        return Answer(self.text, False)
        
class Question:
    
    def __init__(self, text, ans):
        self.text = text
        self.ans = ans
    
    def display(self):
        print("Nội dung câu hỏi: ", self.text)
        for i in self.ans:
            i.display()
    
    def clone(self):
        ans_copy = []
        for i in self.ans:
            ans_copy.append(i.clone())
        return Question(self.text, ans_copy)

    def not_answer(self):
        r = False
        for i in self.ans:
            if i.iscorrect == True:
                r = False
                break
        return r
    
    def correct_answer(self, q):
        r = True
        for i in range(len(self.ans)):
            if self.ans[i].iscorrect != q.ans[i].iscorrect:
                r = False
        return r
    
#Lớp bộ đề
class Test:
    def __init__(self, test):
        self.test = test
        
    def clone(self):
        copy = []
        for i in self.test:
            copy.append(i.clone())
        return Test(copy)

    def compare(self, t):
        correct_answer = 0
        total = len(self.test)
        answered = 0
        not_answered = 0
        for i in range(len(self.test)):
            if self.test[i].not_answer():
                not_answered += 1
            if self.test[i].correct_answer(t.test[i]):
                correct_answer += 1
        return total, total - not_answered, correct_answer
            
    def createThreeQuestion(self):
        answer1 = Answer('Việt Nam', True)
        answer2 = Answer('Mexico', False) 
        answer3 = Answer('Pháp', False) 
        answer4 = Answer('Singapore', True)
        ans1 = [answer1, answer2, answer3, answer4]
        q1 = Question('Nước nào thuộc Châu Á?', ans1)
    
        answer5 = Answer('Lào', False)
        answer6 = Answer('Trung Quốc', False) 
        answer7 = Answer('Việt Nam', True) 
        answer8 = Answer('Malaysia', False)
        ans2 = [answer5, answer6, answer7, answer8]
        q2 = Question('Hà Nội là Thủ đô nước nào?', ans2)
    
        answer9 = Answer('Thái Lan', False)
        answer10 = Answer('Trung Quốc', True) 
        answer11 = Answer('Việt Nam', False) 
        answer12 = Answer('Malaysia', False)
        ans3 = [answer9, answer10, answer11, answer12]
        q3 = Question('Nước nào có diện tích lớn nhất thế giới?', ans3)
        self.test = [q1,q2,q3]    
        
```

```python
from tkinter import *

index = 0
original = Test(None)
original.createThreeQuestion()
finished = False

t = original.clone()

def displayQuestion(i):
    global finished
    question_text.delete(1.0, 'end')
    question_text.insert(1.0, t.test[i].text)
    
    text_answer1.delete(1.0, 'end')
    text_answer1.insert(1.0, t.test[i].ans[0].text)
    cb1.set(t.test[i].ans[0].iscorrect)
    
    text_answer2.delete(1.0, 'end')
    text_answer2.insert(1.0, t.test[i].ans[1].text)
    cb2.set(t.test[i].ans[1].iscorrect)
    
    text_answer3.delete(1.0, 'end')
    text_answer3.insert(1.0, t.test[i].ans[2].text)
    cb3.set(t.test[i].ans[2].iscorrect)
    
    text_answer4.delete(1.0, 'end')
    text_answer4.insert(1.0, t.test[i].ans[3].text)
    cb4.set(t.test[i].ans[3].iscorrect)

    lbFrame_question['text'] = f"Nội dung câu hỏi thứ {index+1}"
    
    if finished == True:
        ckbnt1['state'] = 'disabled'
        ckbnt2['state'] = 'disabled'
        ckbnt3['state'] = 'disabled'
        ckbnt4['state'] = 'disabled'
        ckbnt1['bg'] = 'white'
        ckbnt2['bg'] = 'white'
        ckbnt3['bg'] = 'white'
        ckbnt4['bg'] = 'white'
        text_answer1['bg'] = 'white'
        text_answer2['bg'] = 'white'
        text_answer3['bg'] = 'white'
        text_answer4['bg'] = 'white'
        if t.test[i].correct_answer(original.test[i]):
            for j in range(len(t.test[i].ans)):
                if t.test[i].ans[j].iscorrect:
                    if j == 0:
                        ckbnt1['bg'] = 'green'
                        text_answer1['bg'] = 'green'
                    if j == 1:
                        ckbnt2['bg'] = 'green'
                        text_answer2['bg'] = 'green'
                    if j == 2:
                        ckbnt3['bg'] = 'green'
                        text_answer3['bg'] = 'green'
                    if j == 3:
                        ckbnt4['bg'] = 'green'
                        text_answer4['bg'] = 'green'
        else:
            for j in range(len(t.test[i].ans)):
                if t.test[i].ans[j].iscorrect:
                    if j == 0:
                        ckbnt1['bg'] = 'red'
                        text_answer1['bg'] = 'red'
                    if j == 1:
                        ckbnt2['bg'] = 'red'
                        text_answer2['bg'] = 'red'
                    if j == 2:
                        ckbnt3['bg'] = 'red'
                        text_answer3['bg'] = 'red'
                    if j == 3:
                        ckbnt4['bg'] = 'red'
                        text_answer4['bg'] = 'red'

def goto_next_question(event):
    global index
    if index < len(t.test)-1:
        index += 1
        btn_previous['state'] = 'normal'
        if index == (len(t.test)-1):
            btn_next['state'] = 'disabled'
    displayQuestion(index)   
    
def goto_previous_question(event):
    global index
    if index >0:
        index -= 1
        btn_next['state'] = 'normal'
        if index == 0:
            btn_previous['state'] = 'disabled'
    displayQuestion(index)

def get_temporary_answer1():
    global index
    t.test[index].ans[0].iscorrect = cb1.get()

def get_temporary_answer2():
    global index
    t.test[index].ans[1].iscorrect = cb2.get()

def get_temporary_answer3():
    global index
    t.test[index].ans[2].iscorrect = cb3.get()

def get_temporary_answer4():
    global index
    t.test[index].ans[3].iscorrect = cb4.get()

def completed_test(event):
    global finised
    finised = True
    total, answered, correct_answer = t.compare(original)
    lbResult['text'] = f'Tổng sổ: {total}, trả lời: {answered}, đúng: {correct_answer}, sai: {total - correct_answer}'

    displayQuestion(index)
    
root = Tk()
root.geometry("750x650")
root.resizable(0,0)

lbFrame_question = LabelFrame(master=root, text = 'Nội dung câu hỏi:', font=("Helvetica", 12), fg='blue',height=100)
lbFrame_question.pack(expand=True, fill=BOTH)
question_text = Text(master = lbFrame_question,font=("Helvetica", 12), height=3)
question_text.pack(expand=True, fill=BOTH, padx=10, pady=10)

lbFrame_answer = LabelFrame(root, text="Các đáp án:",font=("Helvetica", 12), fg='blue',height=200)
lbFrame_answer.pack(expand=True, fill=BOTH)

cb1 = BooleanVar()
cb1.set(0)
cb2 = BooleanVar()
cb2.set(0)
cb3 = BooleanVar()
cb3.set(0)
cb4 = BooleanVar()
cb4.set(0)

ckbnt1 = Checkbutton(lbFrame_answer, variable=cb1, text='', onvalue=True, offvalue=False, command=get_temporary_answer1)
ckbnt1.grid(row=0,column=0, padx=5, pady=5)

ckbnt2 = Checkbutton(lbFrame_answer, variable=cb2, text='', onvalue=True, offvalue=False, command=get_temporary_answer2)
ckbnt2.grid(row=0,column=2, padx=5, pady=5)

ckbnt3 = Checkbutton(lbFrame_answer, variable=cb3, text='', onvalue=True, offvalue=False, command=get_temporary_answer3)
ckbnt3.grid(row=1,column=0, padx=5, pady=5)

ckbnt4 = Checkbutton(lbFrame_answer, variable=cb4, text='', onvalue=True, offvalue=False, command=get_temporary_answer4)
ckbnt4.grid(row=1,column=2, padx=5, pady=5)

text_answer1 = Text(lbFrame_answer, font=("Helvetica", 12), width=35, height=8)
text_answer1.grid(row=0,column=1, padx=5, pady=5)

text_answer2 = Text(lbFrame_answer, font=("Helvetica", 12), width=35, height=8)
text_answer2.grid(row=0,column=3, padx=5, pady=5)

text_answer3 = Text(lbFrame_answer, font=("Helvetica", 12), width=35, height=8)
text_answer3.grid(row=1,column=1, padx=5, pady=5)

text_answer4 = Text(lbFrame_answer, font=("Helvetica", 12), width=35, height=8)
text_answer4.grid(row=1,column=3, padx=5, pady=5)

# Thêm các nút bấm thao tác
lbFrame_action = LabelFrame(root, text='Thao tác')
lbFrame_action.pack(expand=True,fill=BOTH)

photo_previous = PhotoImage(file=r'previous.png')
btn_previous = Button(lbFrame_action, text="Câu hỏi trước",font=("Helvetica", 10),image=photo_previous,fg='blue',compound=LEFT)
btn_previous.place(x=10,y=10)
btn_previous.bind("<Button-1>", goto_previous_question)

photo_next = PhotoImage(file=r'next.png')
btn_next = Button(lbFrame_action, text="Câu hỏi tiếp theo",font=("Helvetica", 10),image=photo_next,fg='blue',compound=LEFT)
btn_next.place(x=180,y=10)
btn_next.bind("<Button-1>", goto_next_question)

photo_finish = PhotoImage(file=r'finish.png')
btn_finish = Button(lbFrame_action, text="Kết thúc",font=("Helvetica", 10),image=photo_finish,fg='blue',compound=LEFT)
btn_finish.place(x=350,y=10)
btn_finish.bind("<Button-1>", completed_test)

lbResult = Label(lbFrame_action, text="",font=("Helvetica", 10),fg='green')
lbResult.place(x=500,y=15)

displayQuestion(0)

root.mainloop()
```

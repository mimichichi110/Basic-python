```python
import turtle

tl = turtle.Turtle()
tl.speed(5)

#Snowman head
tl.pencolor('#e0e0e0')
tl.fillcolor('#e0e0e0')
tl.begin_fill()
tl.circle(70,360)
tl.end_fill()

#Snowman body
tl.left(90)
tl.forward(40)
tl.right(90)

tl.left(180)
tl.begin_fill()
tl.circle(100,360)
tl.end_fill()

#left eye
tl.pencolor('black')
tl.fillcolor('black')
tl.penup()
tl.goto(-35, 90)
tl.pendown()
tl.begin_fill()
tl.circle(10,360)
tl.end_fill()

#right eye
tl.penup()
tl.goto(35, 90)
tl.pendown()
tl.begin_fill()
tl.circle(10, 360)
tl.end_fill()

#nose
tl.penup()
tl.goto(0, 70)
tl.pendown()
tl.pencolor('#fda926')
tl.fillcolor('#fda926')
tl.begin_fill()
tl.circle(8, 360)
tl.end_fill()

#mouth
tl.penup()
tl.goto(-20, 40)
tl.pendown()
tl.left(150)
tl.pencolor('black')
tl.pensize(10)
tl.circle(40, 60)

#button
tl.penup()
tl.goto(5, -30)
tl.pendown()
tl.pensize(5)
tl.fillcolor('black')
tl.begin_fill()
tl.circle(10, 360)
tl.end_fill()

tl.penup()
tl.goto(5, -70)
tl.pendown()
tl.pensize(5)
tl.fillcolor('black')
tl.begin_fill()
tl.circle(10, 360)
tl.end_fill()

#left hand
tl.penup()
tl.goto(-100, -35)
tl.pendown()
tl.pencolor('#484d51')
tl.pensize(13)
tl.left(100)
tl.forward(60)

tl.penup()
tl.backward(30)
tl.pendown()
tl.right(70)
tl.forward(23)

#right hand
tl.penup()
tl.goto(100, -35)
tl.pendown()
tl.right(10)
tl.pencolor('#484d51')
tl.pensize(13)
tl.forward(60)

tl.penup()
tl.backward(30)
tl.pendown()
tl.left(70)
tl.forward(23)

#hat
tl.penup()
tl.goto(-70, 120)
tl.pendown()
tl.right(120)
tl.pensize(28)
tl.forward(140)

tl.penup()
tl.left(90)
tl.forward(20)
tl.left(90)
tl.forward(20)
tl.pendown()
tl.pensize(20)
tl.pencolor('#a52719')
tl.forward(100)

tl.penup()
tl.right(90)
tl.forward(10)
tl.right(90)
tl.forward(100)
tl.left(90)
tl.pensize(5)
tl.pencolor('#484d51')
tl.fillcolor('#484d51')
tl.pendown()
tl.begin_fill()
tl.forward(30)
tl.circle(20, 90)
tl.forward(60)
tl.circle(20, 90)
tl.forward(30)
tl.end_fill()




turtle.done()
```

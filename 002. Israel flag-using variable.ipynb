```python
import turtle

tl = turtle.Turtle()
tl.shape('turtle')
tl.hideturtle()
tl.pensize(4)
tl.pencolor('blue')

width = 600
height = 390

tl.penup()
tl.backward(width/2)
tl.right(90)
tl.forward(height/2)
tl.left(90)
tl.pendown()

tl.fillcolor('blue')
tl.begin_fill()
tl.forward(width)
tl.right(90)
tl.forward(height/4)
tl.right(90)
tl.forward(width)
tl.right(90)
tl.forward(height/4)
tl.end_fill()

tl.penup()
tl.forward(height+height/8)
tl.right(90)
tl.pendown()

tl.fillcolor('blue')
tl.begin_fill()
tl.forward(width)
tl.right(90)
tl.forward(height/4)
tl.right(90)
tl.forward(width)
tl.right(90)
tl.forward(height/4)
tl.end_fill()

tl.penup()
tl.backward(height/2)
tl.right(90)
tl.forward(width/3)
tl.pendown()

tl.forward(width/3)
tl.right(120)
tl.forward(width/3)
tl.right(120)
tl.forward(width/3)

tl.right(30)
tl.penup()
tl.forward(height/8)
tl.right(90)
tl.forward(width/6)
tl.right(60)
tl.pendown()

tl.forward(width/3)
tl.right(120)
tl.forward(width/3)
tl.right(120)
tl.forward(width/3)

turtle.done()
```

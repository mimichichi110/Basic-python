```python
from tkinter import *
window = Tk()
window.geometry("300x200")
window.title("Simple calculator")

def add(event):
    try:
        number_1 = int(entry_1.get())
        number_2 = int(entry_2.get())
        lbMessage['text'] = str(number_1) + " + " + str(number_2) + " = " + str(number_1 + number_2)
        lbMessage['fg'] = 'blue'
    except:
        lbMessage['text'] = "Lỗi nhập dữ liệu"
        lbMessage['fg'] = 'red'

def minus(event):
    try:
        number_1 = int(entry_1.get())
        number_2 = int(entry_2.get())
        lbMessage['text'] = str(number_1) + " - " + str(number_2) + " = " + str(number_1 - number_2)
        lbMessage['fg'] = 'blue'
    except:
        lbMessage['text'] = "Lỗi nhập dữ liệu"
        lbMessage['fg'] = 'red'
    
def multiply(event):
    try:
        number_1 = int(entry_1.get())
        number_2 = int(entry_2.get())
        lbMessage['text'] = str(number_1) + " * " + str(number_2) + " = " + str(number_1 * number_2)
        lbMessage['fg'] = 'blue'
    except:
        lbMessage['text'] = "Lỗi nhập dữ liệu"
        lbMessage['fg'] = 'red'
    
def divide(event):
    try:
        number_1 = int(entry_1.get())
        number_2 = int(entry_2.get())
        lbMessage['text'] = str(number_1) + " / " + str(number_2) + " = " + str(number_1 / number_2)
        lbMessage['fg'] = 'blue'
    except:
        lbMessage['text'] = "Lỗi nhập dữ liệu"
        lbMessage['fg'] = 'red'

num1 = Label(window, text = 'Num 1:').place(x=10, y =20)
num2 = Label(window, text = 'Num 2:').place(x=10, y =50)
entry_1 = Entry(window, text = "", width = 20)
entry_1.place(x=70, y =20)
entry_2 = Entry(window, text = "", width = 20)
entry_2.place(x=70, y =50)

bt_add = Button(window, text='+',width = 3, height = 1)
bt_add.place(x=10,y=80)
bt_add.bind("<Button-1>", add)

bt_minus = Button(window, text='-',width = 3, height = 1)
bt_minus.place(x=60,y=80)
bt_minus.bind("<Button-1>", minus)

bt_multiply = Button(window, text='*',width = 3, height = 1)
bt_multiply.place(x=110,y=80)
bt_multiply.bind("<Button-1>", multiply)

bt_divide = Button(window, text='/',width = 3, height = 1)
bt_divide.place(x=160,y=80)
bt_divide.bind("<Button-1>", divide)

out_put = Label(window, text ="Output:").place(x=10,y=120)
lbMessage = Label(window, text = "")
lbMessage.place(x=60,y=120)

window.mainloop()
```
